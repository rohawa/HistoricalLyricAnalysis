```python
#BILLBOARD WEEKLY TOP 5 FROM RB/HIP-HIP GETTER
from __future__ import with_statement
import billboard
import csv
nameSet = []
tempTable = []
oldchart = billboard.ChartData('r-b-hip-hop-songs',date="1975-01-04")
chart = billboard.ChartData('r-b-hip-hop-songs')
with open('test.csv', 'w') as csvfile:
    while chart.previousDate != oldchart.date:
        print(chart.date)
        filewriter = csv.writer(csvfile,
                            quotechar='|', quoting=csv.QUOTE_MINIMAL)
        for x in range(5):
            #print(song)
            song = chart[x]
            filewriter.writerow([chart.date[0:4],song.artist,song.title])
        chart = billboard.ChartData('hot-100', chart.previousDate)
print("finished")
```

```python
#CODE FOR UNIGRAMS
import glob
import errno
import urllib
import re
import nltk
import csv

from nltk.corpus import stopwords
stoplist = stopwords.words('english')

more = ['?',' ','!',';','.',',',';','/','“',
        '’','”','--','I','He','ca','n\'t','do',
        'Do','wo','She','would','could','It','The',
        '(',')',"'","'m",'You',"'s",'And','na',"'ll",
        "'ve","'re",'wan','But',]
stoplist.extend(more)

path = '*.txt'
files = glob.glob(path)
count = 1
year= 1975
alltokens = []

for name in files:
    with open(name) as f:
        alltext = f.read().rstrip()
        alltext = re.sub("\n"," ",alltext)
        tokenized = nltk.word_tokenize(alltext)
        newtok = [x for x in tokenized if x not in stoplist]
        fdist = nltk.FreqDist(newtok)
        with open("unigram"+name[0:4]+".csv", "w") as fp:
            writer = csv.writer(fp, quoting=csv.QUOTE_ALL)
            writer.writerows(fdist.items())
        print(fdist.most_common(20))    
print("finished")
```
